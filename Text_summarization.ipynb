```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import numpy as np
nltk.download('punkt_tab') # Changed 'punkt' to 'punkt_tab' as suggested by the error

# Input text
text = """Natural Language Processing (NLP) is a branch of AI that helps
computers understand human language. It includes tasks such as tokenization,
translation, sentiment analysis, and summarization."""

# Sentence splitting
sentences = nltk.sent_tokenize(text)

# TF-IDF vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

# Sentence similarity scores
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
scores = similarity_matrix.sum(axis=1)

# Select top sentences
N = 2  # number of summary sentences
top_idx = scores.argsort()[-N:]

summary = " ".join([sentences[i] for i in sorted(top_idx)])

print("Summary:\n", summary)
```
