**Step 1: Importing required libraries**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text = """
Shey Phoksundo Lake is one of the most stunning natural landmarks of Nepal, located in the remote Dolpa District within Shey Phoksundo National Park. It is the deepest lake in Nepal and lies at an altitude of approximately 3,611 meters above sea level. The lake is widely known for its crystal-clear turquoise blue water, which appears vibrant due to the reflection of sunlight and the presence of dissolved minerals. Surrounded by steep cliffs, snow-capped peaks, and alpine forests, the lake offers a breathtaking view that attracts trekkers, photographers, and nature lovers from around the world. Near the lake, the Phoksundo Waterfall—one of the highest waterfalls in Nepal—adds to the scenic beauty of the area.

The region around Shey Phoksundo Lake is rich in biodiversity and is home to various rare and endangered species of plants and animals. The national park protects wildlife such as snow leopards, Himalayan blue sheep, musk deer, and different species of birds. Due to its high altitude and cold climate, the ecosystem is fragile and requires careful conservation. The lake itself is oligotrophic, meaning it has low nutrient levels and very clear water, making it environmentally sensitive. For this reason, tourism activities in the area are regulated to minimize environmental damage and preserve the natural balance.

Shey Phoksundo Lake also holds great cultural and religious importance. The nearby villages are mainly inhabited by the Bon and Tibetan Buddhist communities, who have lived in the region for centuries. Ancient monasteries, prayer flags, and traditional stone houses reflect the deep spiritual connection between the local people and nature. The lake is considered sacred, and many religious rituals and festivals are connected to it. Local traditions, language, and lifestyle have remained largely unchanged due to the region’s isolation, making it a living example of Nepal’s cultural diversity.

Despite its beauty and significance, reaching Shey Phoksundo Lake is challenging due to its remote location and difficult terrain. Visitors must trek for several days through rugged landscapes, which makes the journey physically demanding but rewarding. The lake symbolizes the harmony between nature, culture, and spirituality. Preserving Shey Phoksundo Lake is essential not only for environmental reasons but also for protecting Nepal’s natural heritage and cultural identity for future generations.
"""
```

**Step 2: Removing Stop Words and storing them in a separate array of words.**

```python
# Tokenizing the text
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
stopWords = set(stopwords.words("english"))
words = word_tokenize(text)
```

**Step 3: Create a frequency table of words a python dictionary that will keep a record of how many times each word appears in the feedback after removing the stop words.**


```python
# Creating a frequency table to keep the
# score of each word

freqTable = dict()
for word in words:
    word = word.lower()
    if word in stopWords:
        continue
    if word in freqTable:
        freqTable[word] += 1
    else:
        freqTable[word] = 1
```

**Step 4: Assign score to each sentence depending on the words it contains and the frequency table**


```python
# Creating a dictionary to keep the score
# of each sentence
sentences = sent_tokenize(text)
sentenceValue = dict()

for sentence in sentences:
    for word, freq in freqTable.items():
        if word in sentence.lower():
            if sentence in sentenceValue:
                sentenceValue[sentence] += freq
            else:
                sentenceValue[sentence] = freq

```

```python
# Creating a dictionary to keep the score
# of each sentence
sentences = sent_tokenize(text)
sentenceValue = dict()

for sentence in sentences:
    for word, freq in freqTable.items():
        if word in sentence.lower():
            if sentence in sentenceValue:
                sentenceValue[sentence] += freq
            else:
                sentenceValue[sentence] = freq

sumValues = 0
for sentence in sentenceValue:
    sumValues += sentenceValue[sentence]
```

```python
# Average value of a sentence from the original text

average = int(sumValues / len(sentenceValue))
```

```python
# Storing sentences into our summary.
summary = ''
for sentence in sentences:
    if (sentence in sentenceValue) and (sentenceValue[sentence] > (1.2 * average)):
        summary += " " + sentence
print(summary)
```

```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline('summarization')
```

```python
#Function to summarize text
def summarize_text(text, max_length=130, min_length=30, do_sample=False):
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=do_sample)
    return summary[0]['summary_text']
```

```python
# Sample text to summarize
sample_text = """
Rara Lake is the largest lake in Nepal and one of the most beautiful natural attractions in the country. It is located in the remote Mugu District within Rara National Park, at an altitude of about 2,990 meters above sea level. The lake is famous for its deep blue water, which changes color depending on the time of day and weather conditions. Surrounded by pine, spruce, and juniper forests, along with snow-covered hills, Rara Lake offers a peaceful and scenic environment. Due to its untouched beauty and calm atmosphere, it is often described as the “Queen of Lakes” in Nepal.
"""
```

```python
# Summarize the sample text
summarized_text = summarize_text(sample_text)
print(f'Summarized Text:')
summarized_text
```
