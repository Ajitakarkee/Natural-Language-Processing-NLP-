Open the text file for processing:

```python
#Open the text file :
text_file = open("/content/Lab2.txt")

#Read the data :
text = text_file.read()

#Datatype of the data read :
print (type(text))
print("\n")

#Print the text :
print(text)
print("\n")
#Length of the text :
print (len(text))
```

Importing required libraries

```python
#Import required libraries
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
nltk.download ('punkt_tab')
```

Sentence tokenizing:

```python
#Tokenize the text by sentences:
sentences = sent_tokenize(text)

#How many sentences are there? :
print (len(sentences))

#print the sentences:
#print(sentence)
sentences
```

Word tokenizing:

```python
#Tokenize the text with words:
words = word_tokenize(text)
#How many words are there?:
print (len(words))
#Print words:
print (words)
```

Find the frequency distribution:

```python
#Import required libraries :
from nltk.probability import FreqDist
#Find the frequency:
fdist = FreqDist(words)
#Print 10 most common words:
fdist.most_common(10)
```

Plot the frequency graph:

```python
#plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

Remove punctuation marks:

```python
#Empty list to store words:
words_no_punc = []

#Removing punctuation marks:
for w in words:
  if w.isalpha():
    words_no_punc.append(w.lower())
#print the words without punctuation marks:
print (words_no_punc)

print ("\n")

#Length :
print (len(words_no_punc))
```

Displaying words without punctuation

```python
#Frequency distribution :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

Plotting graph without punctuation marks:

```python
#plot the most common words on graph:
fdist.plot(10)
```

Importing stopwords from NLTK

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
#list of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

Removing stopwords:

```python
#Empty list to store clean words :
clean_words = []

for w in words_no_punc:
  if w not in stopwords:
    clean_words.append(w)

print (clean_words)
print("\n")
print (len(clean_words))
```

Final frequency distribution:

```python
#Frequency distribution :
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

Stemming:

Porter’s Stemmer Example 1:

```python
#Stemming Example :

#Import stemming Library :
from nltk.stem import PorterStemmer

porter = PorterStemmer()

#Word-list for stemming :
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(porter.stem(w))
```

Porter’s Stemmer Example 2:

```python
#Stemming Example :

# Import the stemming library
from nltk.stem import PorterStemmer

# Create the stemmer object
porter = PorterStemmer()

# Word-list for stemming
word_list = ["studies", "leaves", "decreases", "plays"]

for w in word_list:
    print(w, "→", porter.stem(w))
```

SnowballStemmer:

```python
#Stemming Example :

#Import stemming Library :
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")

#Word-list for stemming :
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(snowball.stem(w))
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')   # optional but recommended for better lemmas

from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(lemma.lemmatize(w ,pos="v"))
```

Lemmatizer with different POS values

```python
from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
print(lemma.lemmatize('studying' ,pos="v"))
print(lemma.lemmatize('studying' ,pos="n"))
print(lemma.lemmatize('studying' ,pos="a"))
print(lemma.lemmatize('studying' ,pos="r"))
```
