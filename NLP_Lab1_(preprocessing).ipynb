```python
import re
text= "The market extended gains for the seventh consecutive session, climbing 1 percent to end at <b> record </b> closing high on May 31. Reliance Industries <h2> continued to be a leader </h2> in the rally, followed by <br> private banks & financials and FMCG stocks."
text
```

```python
html_remover=re.compile('<.*?>')
text_after_remover=re.sub(html_remover,'',text)
text_after_remover
```

```python
text_data_uppercase = text_after_remover.upper()
text_data_uppercase
```

```python
text_data_lowercase = text_after_remover.lower()
text_data_lowercase
```

```python
text_data= re.sub(r'[^\w\s]','',text_after_remover)
text_data
```

```python
from nltk.tokenize import sent_tokenize
text_data = 'The market extended gains for the seventh consecutive session, climbing 1 percent to end at record closing high on May 31. Reliance Industries continued to be a leader in the rally, followed by private banks & financials and FMCG stocks.'
sent_tokenize(text_data)
```

```python
from nltk.tokenize import word_tokenize
text_data = 'The market extended gains for the seventh consecutive session, climbing 1 percent to end at record closing high on May 31. Reliance Industries continued to be a leader in the rally, followed by private banks & financials and FMCG stocks.'
tokenized_text= word_tokenize(text_data)
print(tokenized_text)
```
